# Data Analysis And Visualizations

```python
cancelled_per = df['is_canceled'].value_counts(normalize = True)  # values_counts give count of freq of a cat. present   normalize give percenatge 
print(cancelled_per)

plt.figure(figsize=(5,4))
plt.title('Resevation Status Count')
plt.bar(['Not Canceled' , 'Canceled'] , df['is_canceled'].value_counts() , edgecolor = 'k', width=0.7)
plt.show()
```

<!-- percentage of is canceled -->

```python
plt.figure(figsize=(8, 4))
ax1 = sns.countplot(
    x='hotel',
    hue='is_canceled',
    data=df,
    palette='Blues'
)

# Legend with proper labels
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(
    title="Cancellation Status",
    labels=["Not Canceled", "Canceled"],
    loc="upper right",
    bbox_to_anchor=(1.25, 1)
)

# Titles and labels
plt.title("Reservation Status in Different Hotels", fontsize=18, weight="bold")
plt.xlabel("Hotel", fontsize=14)
plt.ylabel("Number of Reservations", fontsize=14)

# Add count labels above bars
for p in ax1.patches:
    ax1.annotate(
        str(p.get_height()),
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='bottom',
        fontsize=9,
        color='black',
        xytext=(0, 3),
        textcoords='offset points'
    )

plt.tight_layout()
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
resort_hotel['adr'].plot(figsize=(20,8), label="Resort Hotel", linewidth=2.5)
city_hotel['adr'].plot(label="City Hotel", linewidth=2.5)

plt.title("Average Daily Rate (ADR) in City vs Resort Hotels", fontsize=28, weight="bold")
plt.xlabel("Date", fontsize=18)
plt.ylabel("Average Daily Rate (ADR)", fontsize=18)
plt.legend(fontsize=20)
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(16, 8))

ax1 = sns.countplot(
    x='month_name',       # use month_name for prettier labels
    hue='is_canceled',
    data=df,
    palette='bright',
    order=["January","February","March","April","May","June",
           "July","August","September","October","November","December"]  # keep months in order
)

# Title & labels
plt.title("Reservation Status by Month", fontsize=22, weight="bold")
plt.xlabel("Month", fontsize=16)
plt.ylabel("Number of Reservations", fontsize=16)

# Legend
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(
    title="Cancellation Status",
    labels=["Not Canceled", "Canceled"],
    fontsize=14,
    title_fontsize=14,
    loc="upper right"
)

# Rotate x labels for clarity
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title("Adr", fontsize=28, weight="bold")
plt.bar('month','adr',data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10  countries with reservation canceled')
plt.pie(top_10_country , autopct='%.2f' ,labels= top_10_country.index)
plt.show()
```

```python
df['market_segment'].value_counts()
```

```python
df['market_segment'].value_counts(normalize=True)
```

```python
cancelled_data['market_segment'].value_counts(normalize=True)
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date' , inplace=True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date' , inplace=True)

plt.figure(figsize=(20,3))
plt.title('Avg Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'] , not_cancelled_df_adr['adr'],label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'] , label = 'cancelled')
plt.legend()
```

# Recommendations
Based on the analysis, the following actions are recommended to reduce cancellation rates and improve revenue generation:

- Revise Pricing Strategies: Since cancellation rates rise as the price does, hotels should review their pricing. The City Hotel, in particular, should consider offering discounts or implementing more competitive pricing strategies to reduce the high ADR that correlates with cancellations.

- Targeted Holiday Discounts: The cancellation ratio for the Resort Hotel appears to be higher during peak times like weekends and holidays. Offering reasonable discounts during these periods could incentivize customers to keep their reservations.

- Launch January Marketing Campaigns: Given that January is the month with the highest number of cancellations, hotels should launch targeted marketing campaigns or promotions during this time to increase revenue and mitigate the impact of cancellations.

- Enhance Service Quality for Key Markets: To reduce the high cancellation rate from Portugal, hotels should focus on increasing the quality of their services and overall guest experience for this market segment. This could involve tailored communication, special offers, or improved on-site services.

# Conclusion

This analysis confirms that hotel reservation cancellations are a significant issue driven by several key factors, including high pricing (ADR), seasonality (peaking in January), hotel type (higher in the City Hotel), and customer origin (predominantly Portugal). The data clearly supports the initial hypothesis that price is a major driver of cancellations. By implementing targeted recommendations such as dynamic pricing, seasonal promotions, and enhanced service for key demographics, both City Hotel and Resort Hotel can effectively reduce their cancellation rates, improve room utilization, and ultimately increase their annual revenue.
